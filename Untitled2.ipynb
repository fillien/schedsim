```python
#!/usr/bin/env python

import pandas as pd
import subprocess
import os
import sys
import shutil
from concurrent.futures import ThreadPoolExecutor
from io import StringIO
from datetime import datetime
import matplotlib.pyplot as plt

SCHEDVIEW = "./build/schedview/schedview"
PLATFORM = "./platforms/exynos5422LITTLE.json"

def run_scenario(schedview, platform, logs_path):
    output = subprocess.run(
        [schedview, "--platform", platform, "--directory", logs_path, "--index", "--energy", "--duration"],
        capture_output=True,
        text=True,
        check=True,
    )
    data = StringIO(output.stdout)
    df = pd.read_csv(data, sep=";")
    df["power"] = df["energy"] / df["duration"]
    return df


def process_scheduler(scheduler, platform, directory, log_paths):
    """Process a scheduler's log directory to calculate mean power."""
    return run_scenario(SCHEDVIEW, platform, os.path.join(log_paths[scheduler], directory))["power"].mean()


def process_logs(platform, logs, suffix, output_csv, output_img):
    """Generalized function to process logs for a given test type."""
    utilizations, pa, ffa, csf = [], [], [], []
    log_paths = {
        "pa": f"{logs}_logs_pa_{suffix[0]}",
        "ffa": f"{logs}_logs_ffa_{suffix[1]}",
        "csf": f"{logs}_logs_csf_{suffix[2]}",
    }

    current_util = 0.1

    with ThreadPoolExecutor() as executor:
        for directory in sorted(os.listdir(log_paths["pa"])):
            print(directory)
            utilizations.append(current_util)

            futures = {
                scheduler: executor.submit(process_scheduler, scheduler, platform, directory, log_paths)
                for scheduler in ["pa", "ffa", "csf"]
            }

            pa.append(futures["pa"].result())
            ffa.append(futures["ffa"].result())
            csf.append(futures["csf"].result())

            current_util = round(current_util + 0.1, 2)

    results = pd.DataFrame({
        "util": utilizations,
        "pa": pa,
        "ffa": ffa,
        "csf": csf,
    })

    res = results

    # Plot the data for better understanding
    plt.figure(figsize=(12, 6))
    
    # Plot `ffa` vs `util`
    plt.plot(res['util'], res['pa'], label='pa', marker='o', linestyle='-', color='red')
    plt.plot(res['util'], res['ffa'], label='ffa', marker='o', linestyle='-', color='blue')
    plt.plot(res['util'], res['csf'], label='csf', marker='o', linestyle='-', color='green')
    
    # If other columns are meaningful, they can be plotted as well
    
    # Add labels, title, and legend
    plt.xlabel('Utilization (util)')
    plt.ylabel('FFA')
    plt.title('FFA vs Utilization')
    plt.legend()
    plt.grid(True)
    
    # Display the plot
    plt.show()

    results.to_csv(output_csv, index=False, sep=" ")


def main():
    logs = "data_umax0_9"

    process_logs(
        platform="./platforms/exynos5422LITTLE.json",
        logs=logs,
        suffix=["no_delay", "no_delay", "no_delay"],
        output_csv="energy-no-delay.csv",
        output_img=f"energy_{logs}.png"
    )

    process_logs(
        platform="./platforms/exynos5422LITTLE.json",
        logs=logs,
        suffix=["delay", "delay", "delay"],
        output_csv="energy-delay.csv",
        output_img=f"energy_{logs}.png"
    )

    process_logs(
        platform="./platforms/exynos5422LITTLE.json",
        logs=logs,
        suffix=["delay", "timer", "timer"],
        output_csv="energy-timer.csv",
        output_img=f"energy_{logs}.png"
    )

    process_logs(
        platform="./platforms/exynos5422LITTLEx10.json",
        logs=logs,
        suffix=["no_delay", "no_delay", "no_delay"],
        output_csv="energy-x10-no-delay.csv",
        output_img=f"energy_{logs}.png"
    )

    process_logs(
        platform="./platforms/exynos5422LITTLEx10.json",
        logs=logs,
        suffix=["delay", "delay", "delay"],
        output_csv="energy-x10-delay.csv",
        output_img=f"energy_{logs}.png"
    )

    process_logs(
        platform="./platforms/exynos5422LITTLEx10.json",
        logs=logs,
        suffix=["delay", "timer", "timer"],
        output_csv="energy-x10-timer.csv",
        output_img=f"energy_{logs}.png"
    )
    
```

```python
main()
```
