```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def compute_stats(sched, hw):
    utilizations = []
    mins = []
    maxs = []
    stds = []
    means = []
    varis = []
        
    for i in range(0, 5):
        for j in range(0, 10):
            if ((i < 4 or j <= 0) and ((i > 0) or  (j > 0))):
                filename = "data_" + sched + "_" + str(i) + "_" + str(j) + ".csv"
                freqs = pd.read_csv(filename, sep=";")
                freqs["duration"] = freqs["stop"] - freqs["start"]
                freqs = freqs.drop(freqs[freqs["duration"] < 0.001].index)
                utilizations.append(round(i+j*0.1, 2))
                mins.append(freqs["duration"].min())
                maxs.append(freqs["duration"].max())
                stds.append(freqs["duration"].std())
                means.append(freqs["duration"].mean())
                varis.append(freqs["duration"].var())
    
    results = pd.DataFrame({
        "util": utilizations,
        "min": mins,
        "max": maxs,
        "mean": means,
        "std_dev": stds,
        "var": varis
    })
    outfile = sched + "_" + hw + ".csv"
    results.to_csv(outfile, index=False)
```

# For freqs

```python
compute_stats("pa", "freqs")
compute_stats("ffa", "freqs")
compute_stats("csf", "freqs")
```

```python
data_pa = pd.read_csv("pa_freqs.csv")
data_ffa = pd.read_csv("ffa_freqs.csv")
data_csf = pd.read_csv("csf_freqs.csv")
fig, axs = plt.subplots(3, 1, figsize=(10, 12))

axs[0].plot(data_pa['util'], data_pa['mean'], '-o', data_pa['util'], data_ffa['mean'], '-o', data_csf['util'], data_csf['mean'], '-o')
axs[0].set_title('Mean Duration at Each Utilization Level')
axs[0].set_xlabel('Utilization')
axs[0].set_ylabel('Mean Duration')
axs[0].legend(['pa', 'ffa', 'csf'])
axs[0].grid(True)

axs[1].plot(data_pa['util'], data_pa['std_dev'], '-o', data_pa['util'], data_ffa['std_dev'], '-o', data_csf['util'], data_csf['std_dev'], '-o')
axs[1].set_title('Standard Deviation of Durations at Each Utilization Level')
axs[1].set_xlabel('Utilization')
axs[1].set_ylabel('Standard Deviation')
axs[1].legend(['pa', 'ffa', 'csf'])
axs[1].grid(True)

axs[2].plot(data_pa['util'], data_pa['var'], '-o', data_pa['util'], data_ffa['var'], '-o', data_csf['util'], data_csf['var'], '-o')
axs[2].set_title('Variance of Durations at Each Utilization Level')
axs[2].set_xlabel('Utilization')
axs[2].set_ylabel('Variance')
axs[2].legend(['pa', 'ffa', 'csf'])
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
